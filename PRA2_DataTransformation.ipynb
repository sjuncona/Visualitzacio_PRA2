```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("data/archive/day.csv")
```

```python
df2 = pd.read_csv("data/NYC-BikeShare-2015-2017-combined.csv/NYC-BikeShare-2015-2017-combined.csv")
df2['Datetime'] = pd.to_datetime(df2['Start Time'])
df2 = df2[df2['Datetime'].dt.year == 2016]
```

```python
df3 = pd.read_csv("data/NCDOT_BikePedCrash.csv/NCDOT_BikePedCrash.csv")
```

```python
dict_a1 = {}
for i in range(1,13):
    dict_a1[i] = len(df2[df2['Datetime'].dt.month == i]) 
    
a1 = pd.DataFrame(list(dict_a1.items()),
                  columns = ['Month','Freq']
                 )
a1['year'] = 2016
a1['day'] = 1
a1['Datetime'] = pd.to_datetime(a1[['year','Month','day']])

a1
```

```python
dict_a2 = {}
for i in range(1,13):
    dict_a2[i] = len(df2[(df2['Datetime'].dt.month == i) & (df2['Gender'] == 1)]) 
    
a2 = pd.DataFrame(list(dict_a2.items()),
                  columns = ['Month','Male']
                 )

a2['Female'] = 0
a2['Customer'] = 0
for i in range(1,13):
    a2['Female'].loc[i-1] = len(df2[(df2['Datetime'].dt.month == i) & (df2['Gender'] == 2)]) 
    a2['Customer'].loc[i-1] = len(df2[(df2['Datetime'].dt.month == i) & (df2['Gender'] == 0)]) 
  
a2['year'] = 2016
a2['day'] = 1
a2['Datetime'] = pd.to_datetime(a2[['year','Month','day']])


a2
```

```python
df2['Age_num'] = 2016-df2['Birth Year']
df2['Age'] = 0
df2.loc[df2['Age_num']<=30,'Age']= 'Young' 
df2.loc[(df2['Age_num']> 30) &
        (df2['Age_num']<=60)
        ,'Age']= 'Adult' 
df2.loc[df2['Age_num']>60,'Age']= 'Old' 
```

```python
a3 = pd.DataFrame(columns = ['User','Gender','Age', 'Freq'])
a3.loc[len(a3.index)] = ['Subscriber',
                         'Male',
                         'Young',
                         len(df2[(df2['Gender'] == 1) &
                                 (df2['Age'] == 'Young')
                                ])
                        ]
a3.loc[len(a3.index)] = ['Subscriber',
                         'Male',
                         'Adult',
                         len(df2[(df2['Gender'] == 1) &
                                 (df2['Age'] == 'Adult')
                                ])
                        ]
a3.loc[len(a3.index)] = ['Subscriber',
                         'Male',
                         'Old',
                         len(df2[(df2['Gender'] == 1) &
                                 (df2['Age'] == 'Old')
                                ])
                        ]
a3.loc[len(a3.index)] = ['Subscriber',
                         'Female',
                         'Young',
                         len(df2[(df2['Gender'] == 2) &
                                 (df2['Age'] == 'Young')
                                ])
                        ]
a3.loc[len(a3.index)] = ['Subscriber',
                         'Female',
                         'Adult',
                         len(df2[(df2['Gender'] == 2) &
                                 (df2['Age'] == 'Adult')
                                ])
                        ]

a3.loc[len(a3.index)] = ['Subscriber',
                         'Female',
                         'Old',
                         len(df2[(df2['Gender'] == 2) &
                                 (df2['Age'] == 'Old')
                                ])
                        ]

a3.loc[len(a3.index)] = ['Customer',
                         'Unknown',
                         'Young',
                         len(df2[(df2['Gender'] == 0) &
                                 (df2['Age'] == 'Young')
                                ])
                        ]
a3.loc[len(a3.index)] = ['Customer',
                         'Unknown',
                         'Adult',
                         len(df2[(df2['Gender'] == 0) &
                                 (df2['Age'] == 'Adult')
                                ])
                        ]

a3.loc[len(a3.index)] = ['Customer',
                         'Unknown',
                         'Old',
                         len(df2[(df2['Gender'] == 0) &
                                 (df2['Age'] == 'Old')
                                ])
                        ]
a3
```

```python
a4 = pd.DataFrame(columns=['Weather','Days','Count','Registered','Casual'])

a4.loc[len(a4.index)] = ['Clear, Few clouds, Partly cloudy, Partly cloudy',
                         len(df1[df1['weathersit']==1]),
                         sum(df1.loc[df1['weathersit']==1,'cnt']),
                         sum(df1.loc[df1['weathersit']==1,'registered']),
                         sum(df1.loc[df1['weathersit']==1,'casual'])
                        ]

a4.loc[len(a4.index)] = ['Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist',
                         len(df1[df1['weathersit']==2]),
                         sum(df1.loc[df1['weathersit']==2,'cnt']),
                         sum(df1.loc[df1['weathersit']==2,'registered']),
                         sum(df1.loc[df1['weathersit']==2,'casual'])
                        ]

a4.loc[len(a4.index)] = ['Snow, Thunderstorm , Rain',
                         len(df1[df1['weathersit']==3]),
                         sum(df1.loc[df1['weathersit']==3,'cnt']),
                         sum(df1.loc[df1['weathersit']==3,'registered']),
                         sum(df1.loc[df1['weathersit']==3,'casual'])
                        ]


a4
```

```python
df3.loc[df3['Weather']=='Clear','Weather'] = 'Clear, Few clouds, Partly cloudy, Partly cloudy'

df3.loc[df3['Weather']=='Cloudy','Weather'] = 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist'
df3.loc[df3['Weather']=='Fog, Smog, Smoke','Weather'] = 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist'

df3.loc[df3['Weather']=='Rain','Weather'] = 'Snow, Thunderstorm , Rain'
df3.loc[df3['Weather']=='Snow, Sleet, Hail, Freezing Rain/Drizzle','Weather'] = 'Snow, Thunderstorm , Rain'
df3.loc[df3['Weather']=='Other','Weather'] = 'Snow, Thunderstorm , Rain'


set(df3['Weather'])
```

```python
months = ['January','February','March','April','May','June','July','August','September','October','November','December']
weather = ['Clear, Few clouds, Partly cloudy, Partly cloudy','Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist','Snow, Thunderstorm , Rain']
```

```python
a5 = pd.DataFrame(columns=['Month','Weather','Freq'])

for mes in months:
    for temps in weather:
        a5.loc[len(a5.index)] = [mes, 
                                 temps,
                                 len(df3[(df3['CrashMonth']==mes) &
                                         (df3['Weather']==temps) 
                                        ])
                                ]
        
a5
```

```python
ratio_dona = len(df2[df2['Gender']==2])/(len(df2[df2['Gender']==2])+len(df2[df2['Gender']==1]))
ratio_adult = len(df2[df2['Age']=='Adult'])/len(df2)
ratio_jove = len(df2[df2['Age']=='Young'])/len(df2)
ratio_vell = len(df2[df2['Age']=='Old'])/len(df2)
```

```python
df3['BikeAge'] = df3['BikeAge'].str.replace('+', '', regex=False).copy()
```

```python
a6 = pd.DataFrame(columns=['Bikes Rented','Accidents','Name','C'])

for i in range(1,13):
    a6.loc[len(a6.index)] = [6*sum(df1.loc[df1['mnth']==i,'cnt']), 
                             len(df3[df3['CrashMonth']==months[i-1]]),
                             months[i-1],
                             'Time'
                            ]

for i in range(1,4):
    a6.loc[len(a6.index)] = [6*sum(df1.loc[df1['weathersit']==i,'cnt']), 
                             len(df3[df3['Weather']==weather[i-1]]),
                             weather[i-1],
                             'Weather'
                            ]
    
a6.loc[len(a6.index)] = [int(6*sum(df1['cnt'])*(1-ratio_dona)), 
                         len(df3[df3['BikeSex']=='Male']),
                         'Male',
                         'Gender'
                        ]

a6.loc[len(a6.index)] = [int(6*sum(df1['cnt'])*(ratio_dona)), 
                         len(df3[df3['BikeSex']=='Female']),
                         'Female',
                         'Gender'
                        ]

a6.loc[len(a6.index)] = [int(6*sum(df1['cnt'])*(ratio_dona)), 
                         len(df3[df3['BikeSex']=='Female']),
                         'Female',
                         'Gender'
                        ]
a6.loc[len(a6.index)] = [int(6*sum(df1['cnt'])*(ratio_jove)), 
                         len(df3[df3['BikeAge'].astype(int)<=30]),
                         'Young',
                         'Age'
                        ]

a6.loc[len(a6.index)] = [int(6*sum(df1['cnt'])*(ratio_adult)), 
                         len(df3[(df3['BikeAge'].astype(int)>30) &
                                 (df3['BikeAge'].astype(int)<=60)
                                ]),
                         'Adult',
                         'Age'
                        ]

a6.loc[len(a6.index)] = [int(6*sum(df1['cnt'])*(ratio_vell)), 
                         len(df3[df3['BikeAge'].astype(int)> 60]),
                         'Old',
                         'Age'
                        ]


a6
```
